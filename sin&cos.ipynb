```python
class Man:
    """functions of class """

    def __init__(self, name):
        self.name = name
        

    def sin(self):
        import numpy as np
        import matplotlib.pyplot as plt
        x = np.arange(0, 6, 0.1)
        y1 = np.sin(x)
        plt.plot(x, y1, label="sin")
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.title('sin')
        plt.legend()
        plt.show()
    def cos(self):
        import numpy as np
        import matplotlib.pyplot as plt
        x = np.arange(0, 6, 0.1)
        y2 = np.cos(x)
        plt.plot(x, y2, label="cos")
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.title('cos')
        plt.legend()
        plt.show()
```

```python
m = Man("David")
m.sin()
m.cos()
```
